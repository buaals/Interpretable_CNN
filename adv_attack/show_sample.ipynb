```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# FGSM data
with open('fgsm_data.pkl','rb') as fr:
    fgsm_data=pickle.load(fr)
with open('bim_data.pkl','rb') as fr:
    bim_data=pickle.load(fr)
with open('DeepFool_data.pkl','rb') as fr:
    DF_data=pickle.load(fr)
```

```python
print fgsm_data.shape
print bim_data.shape
print DF_data.shape
fgsm_data=fgsm_data[:,:,:,0]
bim_data=bim_data[:,:,:,0]
DF_data=DF_data[:,:,:,0]
from keras.datasets import mnist
```

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
test_images = test_images.reshape(-1, 28, 28, 1)
test_images=np.pad(test_images,  ((0,0),(2,2),(2,2),(0,0)), mode='constant')
test_images=test_images[:,:,:,0]
print test_images.shape
```

```python
fig, axes=plt.subplots(4,4,figsize=(20,20))
fig.tight_layout()
axes[0][0].imshow(test_images[0])
axes[0][0].set_title('ORIGIN')
axes[0][1].imshow(fgsm_data[0])
axes[0][1].set_title('FGSM')
axes[0][2].imshow(bim_data[0])
axes[0][2].set_title('BIM')
axes[0][3].imshow(DF_data[0])
axes[0][3].set_title('Deep Fool')

axes[1][0].imshow(test_images[1])
axes[1][1].imshow(fgsm_data[1])
axes[1][2].imshow(bim_data[1])
axes[1][3].imshow(DF_data[1])

axes[2][0].imshow(test_images[2])
axes[2][1].imshow(fgsm_data[2])
axes[2][2].imshow(bim_data[2])
axes[2][3].imshow(DF_data[2])

axes[3][0].imshow(test_images[3])
axes[3][1].imshow(fgsm_data[3])
axes[3][2].imshow(bim_data[3])
axes[3][3].imshow(DF_data[3])
plt.savefig('sample.png')
plt.show()
```
